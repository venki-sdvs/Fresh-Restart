```python
# To read the dataset from the folder path mentioned above
df = pd.read_csv('Salary_Data.csv')

# To check the first 5 rows of the dataset created
df.head()
```

```python
# To check the shape of the dataset
df.shape
```

```python
# To check the columns in the dataset
df.columns
```

```python
# create new datasets, by retaining input and output features separately 

X = df[['YearsExperience']] # input feature
y = df[['Salary']] #output feature
```

```python
# Import the library from sklearn to split the X, y into train and test

from sklearn.model_selection import train_test_split
```

```python
# Split the X , y to train and test- test_size refers the percentage of the split and random_state refers to reproducability of the same output

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =0)
```

```python
# To know the shape of the newly created datasets : Train & Test
print(r'Shape of X_train:',X_train.shape)
print(r'Shape of X_test:',X_test.shape)

print(r'Shape of y_train:',y_train.shape)
print(r'Shape of y_test:',y_test.shape)
```

```python
# Import the regression model from the library linear_model

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train, y_train) # model trainoing happens here

```

```python
# We can view and check the values of coffiecient ( Weights) and Biases(Intercept)

print(regressor.coef_)
print(regressor.intercept_)
```

```python
# To check how the model has been learnt 
y_pred = regressor.predict(X_test)
```

```python
# import the evaluation metrics to evalue the model created
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error
```

```python
# Lets check and evaluate the model using evaluation metrics
# R_Square is close to 1 means the model is good and if the model is close to 0(Zero) - its a bad model

r_square = r2_score(y_test, y_pred)
print(r_square)
rmse =  root_mean_squared_error(y_test,y_pred)
print(rmse)
mae = mean_absolute_error(y_test, y_pred)
print(mae)
```

```python
# The packaging of the model for future use and for deployment
import pickle
filename = 'finalized_linearReg.sav'

pickle.dump(regressor, open(filename, 'wb')) 
```

```python
# unpack the model for the usage
loaded_model = pickle.load(open('finalized_linearReg.sav', 'rb'))
```

```python
# To check the unpackaged model performance with given input value
result=loaded_model.predict([[13]])
print(result)
```
